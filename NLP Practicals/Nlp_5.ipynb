**Morphology is the study of the way words are built up from smaller meaning bearing units.
Study and understand the concepts of morphology by the use of add delete table.**

**Function to Apply Morphological Rules**


*   Takes a base word, a morpheme to add, and a morpheme to delete.
*   Removes the specified characters
*   Adds the morpheme
*   Returns the transformed word.

```python
from tabulate import tabulate

def apply_morphology(base_word, add=None, delete=None):
    """Applies morphological transformation by adding or deleting morphemes."""
    if delete:
        transformed_word = base_word.replace(delete, "")  # Remove specified characters
    else:
        transformed_word = base_word

    if add:
        transformed_word += add  # Add suffix/prefix

    return transformed_word

# Define words with morphological changes
morphology_rules = [
    ("play", "ing", None),
    ("run", "ning", None),
    ("happy", "ly", None),
    ("agree", "ment", None),
    ("go", "ne", "o"),
    ("write", "er", None),
    ("big", "ger", None),
    ("child", "ren", "d")
]

# Generate results
table_data = []
for base, add, delete in morphology_rules:
    result = apply_morphology(base, add, delete)
    table_data.append([base, add if add else "-", delete if delete else "-", result])

# Display as a table
headers = ["Base Word", "Add", "Delete", "Result"]
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```
